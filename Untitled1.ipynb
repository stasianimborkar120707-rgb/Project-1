```python

# ==========================================
# MNIST Digit Classification Neural Network
# ==========================================

# [span_1](start_span)Step 1: Import Required Libraries[span_1](end_span)
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# [span_2](start_span)Step 2: Load the MNIST Dataset[span_2](end_span)
# The dataset is built into Keras for easy access
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# [span_3](start_span)Step 3: Understand Dataset Shape[span_3](end_span)
print(f"Training data shape: {X_train.shape}") # 60,000 images of 28x28
print(f"Test data shape: {X_test.shape}")     # 10,000 images of 28x28

# [span_4](start_span)Step 4: Visualize Sample Image[span_4](end_span)
plt.imshow(X_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.show()

# [span_5](start_span)Step 5: Normalize the Dataset[span_5](end_span)
# Scaling pixel values from 0-255 to 0-1 for better performance
X_train = X_train / 255.0
X_test = X_test / 255.0

# [span_6](start_span)Step 6: Build Neural Network Model[span_6](end_span)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),    # Converts 2D image to 1D vector
    layers.Dense(128, activation='relu'),    # Hidden layer with ReLU
    layers.Dense(10, activation='softmax')   # Output layer for 10 digits (0-9)
])

# [span_7](start_span)Step 7: Compile the Model[span_7](end_span)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# [span_8](start_span)Step 8: Train the Model[span_8](end_span)
print("\nStarting Training...")
model.fit(X_train, y_train, epochs=10)

# [span_9](start_span)Step 9: Evaluate the Model[span_9](end_span)
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"\nTest Accuracy: {test_acc*100:.2f}%")

# [span_10](start_span)Step 10: Make Predictions[span_10](end_span)
predictions = model.predict(X_test)

# [span_11](start_span)Step 11 & 12: Predict a Single Image and Display Result[span_11](end_span)
index = 0 # Change this index to see different predictions
predicted_label = np.argmax(predictions[index])
actual_label = y_test[index]

plt.imshow(X_test[index], cmap='gray')
plt.title(f"Actual: {actual_label} | Predicted: {predicted_label}")
plt.show()

print(f"The model predicts this digit is a: {predicted_label}")
```
